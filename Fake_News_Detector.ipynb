```python
import numpy as np
import pandas as pd
```

```python
true = pd.read_csv('/content/True.csv')
fake = pd.read_csv('/content/Fake.csv')
```

```python
true.head()
```

```python
fake.head()
```

```python
true['label'] = 1
fake['label'] = 0
```

```python
true.head()
```

```python
news = pd.concat([fake,true],axis=0)
```

```python
news.head()
```

```python
news.tail()
```

```python
news.isnull().sum()
```

```python
news = news.drop(['title','subject','date'],axis=1)
```

```python
news.head()
```

```python
news = news.sample(frac=1)
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'],axis=1,inplace =True)
```

```python
news
```

```python
import re
```

```python
def wordconverter(text):
  text = str(text)
  text = text.lower()
  text = re.sub(r'https?://\S+|www\.\S+','',text)
  text = re.sub(r'<.*?>','',text)
  text = re.sub(r'[^\w\s]','',text)
  text = re.sub(r'\d','',text)
  text = re.sub(r'\n',' ',text)
  return text
```

```python
news['text'] = news['text'].apply(wordconverter)
```

```python
news
```

```python
x = news['text']
y = news['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
LR.fit(xv_train,y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(xv_train,y_train)
```

```python
pred_dtc = DTC.predict(xv_test)
```

```python
DTC.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xv_train,y_train)
```

```python
predict_rfc = rfc.predict(xv_test)
```

```python
rfc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,predict_rfc))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier()
```

```python
gbc.fit(xv_train,y_train)
```

```python
pred_gbc = gbc.predict(xv_test)
```

```python
gbc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_gbc))
```

```python
def output_label(n):
  if n==0:
    return "It's a Fake news"
  elif n==1:
    return "It's a Genuine news"
```

```python
def manual_testing(news):
  if not isinstance(news,str):
    raise ValueError("News article must be a string")
  testing_news = {"text":[news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test.apply(wordconverter)
  new_x_test = new_def_test["text"]
  new_xv_test = vectorization.transform(new_x_test.fillna(' '))
  pred_lr = LR.predict(new_xv_test)
  # pred_dtc = dtc.predict(new_xv_test)
  pred_gbc = gbc.predict(new_xv_test)
  pred_rfc = rfc.predict(new_xv_test)
  return "\n\nLR Prediction: {}  \nGBC Prediction: {}  \nRFC Prediciton: {}".format(
      output_label(pred_lr[0]),output_label(pred_gbc[0]),output_label(pred_rfc[0])
      )
```

```python
news_article = str(input("Enter the news article: "))
```

```python
manual_testing(news_article)
```
